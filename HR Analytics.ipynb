# CH2.2 Data Cleaning & Aggregation


```python
# Cleaning Up Categorical Values
df_general['Attrition'].value_counts()
```

```python
# Grouping Similar Categorical into One
df_general['Attrition'] = df_general['Attrition'].replace('Y','Yes')
df_general['Attrition'] = df_general['Attrition'].replace('N','No')
```

```python
df_general['Attrition'].value_counts()
```

```python
# Converting Categorical Values to Boolean Values
df_general['Attrition'] = df_general['Attrition'] == 'Yes'
```

```python
df_general['Attrition'].value_counts()
```

```python
df_general['Department'].value_counts()
```

```python
df_general['Department'] = df_general['Department'].str.replace("_", " ", regex=True)
df_general['Department'] = df_general['Department'].str.replace("HR", "Human Resources", regex=True)
```

```python
df_general['Department'].value_counts()
```

```python
# Converting Complex Dataset to Usable Dataset
```

```python
df_in_time = pd.read_csv('in_time.csv',index_col=0)
df_out_time = pd.read_csv('out_time.csv',index_col=0)
```

```python
df_in_time.head()

# column name is the date of work and values is the work start time
# Null value most likely represents PTO or sick day
```

```python
df_in_time[['2015-01-01','2015-01-05','2015-01-02']].info()

# Currently the dataset is in object data type. So, we need to first convert them into datetime type.
```

```python
df_in_time = df_in_time.apply(pd.to_datetime, errors='coerce')
df_out_time = df_out_time.apply(pd.to_datetime, errors='coerce')
```

```python
df_in_time[['2015-01-01','2015-01-05','2015-01-02']].info()
```

```python
# Now that all the values are converted to datetime values, we can collect them into usuable form.
# There could be many different analytics that we can perform: daily work hours, number of PTO/sick day, etc.
```

```python
# Collecting daily work hour of each employees
df_work_hours = pd.DataFrame()

dates = df_in_time.columns # all dates in the dataset

for date in dates:
    df_work_hours[date] = df_out_time[date] - df_in_time[date] # calculating daily work hour
```

```python
df_work_hours.head()
```

```python
df_work_hours = df_work_hours.T #Transposing the dataframe to 
df_work_hours
```

```python
# Aggregating Daily Work Hour into Average Daily Work Hour Per Employee
average_work_hour = df_work_hours.mean()
average_work_hour =  average_work_hour.dt.total_seconds() / 3600
average_work_hour = average_work_hour.round(2)
average_work_hour
```

```python
annual_work_days = df_work_hours.count()
annual_work_days
```

```python
# Adding calculated values into df_general Dataset
df_general['avg_work_hour'] = np.round(df_general['EmployeeID'].map(average_work_hour)*2, 0)/2
df_general['annual_work_days'] = df_general['EmployeeID'].map(annual_work_days)
```

```python
# Merging the rest of the Dataset using pd.merge() function
df_comb = pd.merge(df_general, df_e, how='inner', on='EmployeeID')
df_comb = pd.merge(df_comb, df_m, how='inner', on='EmployeeID')
df_comb.head()
```

# Ch 5 Drawing Conclusion & Recommendation

### Summary of Findings:
Data analysis of employee turnover revealed key differences between employees who left and those who stayed:
- Lower job satisfaction and lower monthly income were associated with higher turnover.
- Employees who left had longer working hours and more time since their last promotion.
- A statistically significant difference was observed in performance ratings between those who stayed and those who left.

### Recommendations for Improvement:
1. Enhance Job Satisfaction & Compensation:
    - Conduct employee satisfaction surveys to identify key areas of dissatisfaction.
    - Review and adjust compensation structures to ensure competitive salaries aligned with market benchmarks.
    - Offer non-monetary benefits (e.g., flexible work arrangements, wellness programs, career development opportunities).

2. Manage Workload and Work Hours:

    - Implement workload-balancing strategies to prevent excessive overtime.
    - Encourage efficient task management and delegation to reduce employee burnout.
    - Promote a culture of work-life balance through flexible scheduling and remote work options where feasible.

3. Improve Career Growth Opportunities:
    - Establish a clear promotion pathway and communicate career progression opportunities transparently.
    - Implement regular performance and promotion reviews to recognize and reward high-performing employees.
    - Provide structured mentorship and leadership programs to support employee development.

4. Address Performance Rating Discrepancies:
    - Review performance evaluation criteria to ensure fairness and consistency.
    - Provide training for managers on unbiased performance assessments and feedback delivery.
    - Offer personalized development plans for employees who need improvement to retain talent.

### Conclusion:
By addressing job satisfaction, compensation, work hours, career progression, and performance evaluation fairness, the company can reduce turnover rates, improve employee engagement, and enhance overall retention. Implementing these recommendations will lead to a more stable and productive workforce, ultimately benefiting the organization’s long-term success.
